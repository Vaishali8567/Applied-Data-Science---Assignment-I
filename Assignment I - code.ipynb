```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Read the dataset from CSV
df = pd.read_csv("Assignment I.csv")

# Display the first few rows of the dataframe
print(df.head())

# Display summary statistics of the numerical columns
print(df.describe())

# Select only numeric columns
numeric_df = df.select_dtypes(include=[np.number])

# Display correlation matrix
print(numeric_df.corr())
```

```python
# Plot a histogram of Happiness Score
plt.figure(figsize=(10, 6))
plt.hist(df['Happiness Score'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Happiness Score')
plt.ylabel('Frequency')
plt.title('Distribution of Happiness Score')
plt.grid(True)
plt.show()
```

```python
# Plot a scatter plot of Economy (GDP per Capita) vs Happiness Score
plt.figure(figsize=(10, 6))
plt.scatter(df['Economy (GDP per Capita)'], df['Happiness Score'], color='orange')
plt.xlabel('Economy (GDP per Capita)')
plt.ylabel('Happiness Score')
plt.title('Economy (GDP per Capita) vs Happiness Score')
plt.grid(True)
plt.show()
```

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include='number')

# Plot a heatmap of correlation matrix
plt.figure(figsize=(10, 6))
plt.imshow(numeric_df.corr(), cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.title('Correlation Matrix Heatmap')
plt.xticks(range(len(numeric_df.columns)), numeric_df.columns, rotation=90)
plt.yticks(range(len(numeric_df.columns)), numeric_df.columns)
plt.show()
```
